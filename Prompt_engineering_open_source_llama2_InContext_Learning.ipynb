##Loading llama27b

```python
from transformers import AutoTokenizer
import transformers
import torch

#specifying the model we want to use
model = "meta-llama/Llama-2-7b-chat-hf"
```

```python
#loading the pretrained tokenizer
tokenizer = AutoTokenizer.from_pretrained(model)
```

```python
#making it ready for the text geneartion
pipeline=transformers.pipeline("text-generation",
                               model=model,
                               torch_dtype=torch.float16,
                               device_map="auto")
```

#### Defining a function to accept a prompt and return the response

```python
def get_response(prompt):

  sequences = pipeline(prompt,
      #do_sample=True,
      return_full_text=False,
      top_k=10,
      num_return_sequences=1,
      eos_token_id=tokenizer.eos_token_id,
      max_length=2000,)

  return sequences[0]['generated_text']
```

```python
### feeding this information to llm
prompt = f'''

<s>[INST] <<SYS>>
You are a helpful, respectful and honest assistant. Always answer as helpfully as possible, while being safe.
Your answers should not include any harmful, unethical, racist, sexist, toxic, dangerous, or illegal content.
<</SYS>>
What are the type of posts available  given in the text in delimeters <> :  <{text_1}>
[/INST]

'''
print(get_response(prompt))
```

```python
### creating a question and answer function for in context learning for llms
def get_answer(question,context_text):
  prompt = f'''

  <s>[INST] <<SYS>>
  You are a helpful, respectful and honest assistant. Always answer as helpfully as possible, while being safe.
  Your answers should not include any harmful, unethical, racist, sexist, toxic, dangerous, or illegal content.
  <</SYS>>
  {question}  given in the text in delimeters <> :  <{context_text}>
  [/INST]

  '''
  print(get_response(prompt))
```
